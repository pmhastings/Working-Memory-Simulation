## The model

```python
U_0 = 0
U_ex = 7
U_in = -1
U_K = -1

# theta = 1

tau_U = 2.5 * ms
tau_theta = 10 * ms
tau_G_K = 5 * ms

# TODO: CHECK THESE: The "time constants" on p, 770, rt, halfway down?
tau_e = 1 * ms           # for most places
tau_e_pf_it = 7.5 * ms   # pf to it
tau_i = 1.5 * ms         # inhibitory connections

w_ext = 0.8           # p. 770, 2nd col, line 16, JUST FOR INPUTS

# Back to values from the paper
w_intra_it = 0.0125 # (0.006)   # p. 770, 2nd, line 20
w_intra_pf = 0.005  # (0.0025)
w_pf_it = 0.005     # (0.0025)
w_chunking = w_intra_it / 15 
w_inh = 0.005       # (0.0025)  # p. 770, 2nd col, line 32ish

# Delays in paper:
del_intra = '(.5+rand())*ms' # p. 768: intra_IT: rand(0.5 - 1.5 ms)
del_chunk = 1*ms  # p. 768: chunking: 1 ms
del_pf_it = 12*ms # p. 769: PF <-> IT: 12 ms
# p. 769: intra_PF: rand(0.5 - 1.5 ms) As above (del_intra)

# This was in: SIP, chunking, and inhibitory inter-assembly, 
# probably because that's what it says on p. 770 about "time constants".
# But 
# del_intra = 1*ms      # "time constants", p 770, but not sure about this
# This was in: SPI
# I think del_fdbk is for PF -> IT
# del_fdbk = 12*ms  # p. 769, fig 1 and 2nd full para
# del_fdbk = 7.5*ms # p. 770, halfway down r col
# but I put that in tau_e instead (uncertain about this)

# 7.5 for indiv spikes, but bursts at 6 and 5 (p. 780)
b = 7.5  # 5 # 7

# Number of steps it takes for U to drop below threshold
# Now I make sure that RT is always = 1.
RT = 1
b_step = b / (defaultclock.dt*RT)

# size of assemblies
pfsz = 50
itsz = 100
# number of assemblies
nasses = 20
```

```python
# Equations 
# For this one, I'm having three different G_e's for different tau's

# For theta 
c = 0.2
theta_0 = 1

# Adding an "active" parameter

eqs = '''
# Equation 1
# With extra term for slower excitatory feedback connection PF -> IT
# dU/dt = (-(U - U_0) + floor(G_K * (U_K - U) + (G_e + G_e_pf_it) * (U_ex - U) + G_i*(U_in - U)))/tau_U: 1
dU/dt = (-(U - U_0) + G_K * (U_K - U) + (G_e + G_e_pf_it) * (U_ex - U) + G_i*(U_in - U))/tau_U: 1

# Equation 2 is for theta, but they leave at 1 (except for Fig 2???)
dtheta/dt = (-(theta - theta_0) + c*U)/tau_theta : 1

# Equation 3: Potassium current
# Now adopting single-step update of G_K to b
# dG_K/dt = -G_K / tau_G_K + b_step*S: 1
dG_K/dt = -G_K / tau_G_K : 1

# Equation 4: IS THIS RIGHT???
# Hack: if U>theta, then 1, else 0
# S = int(U>theta) : 1
active : boolean
S : 1

# Equations 5a and 5b
# This is for the magic of excitatory and inhibitory PSPs 
# (for inter-assembly connections),
# according to Brian's converting to ODE page.
# I'm guessing that I should have the big G here, not little g
# tau_exc : second

dG_e/dt = -G_e/tau_e : 1
dG_e_pf_it/dt = -G_e_pf_it/tau_e_pf_it : 1
dG_i/dt = -G_i/tau_i : 1
'''
```

```python
start_scope()

# PF and IT Neuron Assemblies

# de/activation scheme:
# U >= threshold and active and S=0: S = 1, G_K = b, active=false
# S==1: S = 0
# U < threshold and not active: active = true
# 2018 11 02: Just realized that these were firing a lot more than
# I wanted because firing was just based on 'U>=theta'
PFG = NeuronGroup(pfsz*nasses, eqs, threshold='(U>=theta) and active and (S<.5)', method='euler', 
                    events={'resetS': 'S > .5',
                            'reactivate': '(U < theta) and not(active)',
                            'setS': '(U >= theta) and active and (S < .5)'})
ITG = NeuronGroup(itsz*nasses, eqs, threshold='(U>=theta) and active and (S<.5)', method='euler',
                    events={'resetS': 'S > .5',
                            'reactivate': '(U < theta) and not(active)',
                            'setS': '(U >= theta) and active and (S < .5)'})
PFG.theta = 1.0
ITG.theta = 1.0
PFG.active = True
ITG.active = True
PFG.run_on_event('resetS', 'S = 0')
PFG.run_on_event('reactivate', 'active = True')
PFG.run_on_event('setS', 'S = 1; G_K = b; active=False') 
ITG.run_on_event('resetS', 'S = 0')
ITG.run_on_event('reactivate', 'active = True')
ITG.run_on_event('setS', 'S = 1; G_K = b; active=False') 

# Make subgroups (= assemblies)
PFs = [PFG[k*pfsz:(k+1)*pfsz] for k in range(nasses)]
ITs = [ITG[k*itsz:(k+1)*itsz] for k in range(nasses)]

net = Network(collect())

# Make connections

# Intra-assembly 
for ng in PFs:
    SP = Synapses(ng, ng, on_pre = 'G_e += w_intra_pf')
    SP.connect(condition='i!=j')
    SP.delay = del_intra
    net.add(SP)

for ng in ITs:
    SI = Synapses(ng, ng, on_pre = 'G_e += w_intra_it')
    SI.connect(condition='i!=j')
    SI.delay = del_intra 
    net.add(SI)

# PF to and from corresponding IT
for it, pf in zip(ITs, PFs):
    SIP = Synapses(it, pf, on_pre='G_e += w_pf_it', delay = del_pf_it)
    SIP.connect()
    # different G_e
    SPI = Synapses(pf, it, on_pre='G_e_pf_it += w_pf_it', delay = del_pf_it)  
    SPI.connect()
    net.add(SIP)
    net.add(SPI)

# For alpha synapses, I might need to:
# make the weight a parameter AND/OR
# set it to be clock-driven, or summed AND/OR
# set the method of computation (euler)

# Chunking fields: Synchronizing intra-object connections in IT
# Connect each assembly to all other assemblies within the same object/chunk
# Delay = 1 ms
for k in range(5):
    for m in range(4):
        for n in range(4):
            if m!=n:
                frm = ITs[k*4+m]
                to = ITs[k*4+n]
                S = Synapses(frm, to, on_pre='G_e += w_chunking', delay= del_chunk)
                S.connect()
                net.add(S)
            
# Inhibitory assemblies
# Paper seems ambiguous about how this is done: "implicitly modeled", "hypothetical assemblies"

# Implicit way: Each neuron of each assembly connected to each neuron of 
# 16 competing-object assemblies: 1600 incoming connections to each 
# neuron = 160,000 per assembly

# Explicit way: 100 (?) neurons in assembly, each of which have 1600 incoming 
# connections + completely connected between inhibitory assembly and IT assembly 
# = 160,000 + 10,000 = 170,000, so not a radical difference in num of connections.

# I'm going with implicit, because I think that's what they really meant.
# Connect from all others that are not within the object
# (What makes this inhibitory?  U_inh is -1 (like U_K).)
for k, it_to in enumerate(ITs):
    for m, it_frm in enumerate(ITs):
        # Exclude the ones that are in the same "object"
        k_ex = (k // 4)*4
        # if not within the same chunk, make connections
        if not(k_ex <= m < (k_ex+4)):
            # not specifying tau here, because it's same for all inhib
            # I think this is the same delay
            S = Synapses(it_frm, it_to, on_pre='G_i += w_inh', delay=del_chunk)
            S.connect()
            net.add(S)

#######################
# INPUT

r_0_it = 20*Hz   # p = .01 = 1000 ms/s * .01/.5 ms = 20 / s = 20 Hz
r_stim = 200*Hz  # p = .1 
r_0_pf = 10*Hz   # p = .005 (bkgrnd for PF)

in_bkgrnd = PoissonInput(PFG, 'G_e', 1, r_0_pf, w_ext)
net.add(in_bkgrnd)

# for 500 ms run, stim on from 100-200
ta_stim = TimedArray([0,w_ext,0,0,0], dt=100*ms)
ta_rest = TimedArray([w_ext,0,w_ext,w_ext,w_ext], dt=100*ms)
in_stim = PoissonInput(ITG, 'G_e', 1, r_stim, 'ta_stim(t)')
in_rest = PoissonInput(ITG, 'G_e', 1, r_0_it, 'ta_rest(t)')
net.add(in_stim)
net.add(in_rest)

# collect()
# Only includes objects that have a name, not ones in collections.

#######################
# Aggregate measure of activity
# M = SpikeMonitor(ITG)

# Brunel style
PRMs = [PopulationRateMonitor(ng) for ng in ITs]
net.add(PRMs)

SM = SpikeMonitor(ITG)
net.add(SM)

StM = StateMonitor(ITG, ('U', 'theta', 'G_K', 'G_e', 'G_e_pf_it', 'G_i', 'S'), record=True)
net.add(StM)

net.store()
```

```python
net.restore()

net.run(500*ms, report='stdout')
```

```python
def make_ui(wit, wip, wpi, wi, wch):
    siit = widgets.FloatSlider(description='intra IT:', min=0,max=0.075,step=0.0005, 
                               value=wit, readout_format='.4f')
    sipf = widgets.FloatSlider(description='intra PF:', min=0,max=0.075,step=0.0005, 
                               value=wip, readout_format='.4f')
    spi = widgets.FloatSlider(description='PF to IT:', min=0,max=0.075,step=0.0005, 
                               value=wpi, readout_format='.4f')
    sinh = widgets.FloatSlider(description='inhibit:', min=0,max=0.075,step=0.0005, 
                               value=wi, readout_format='.4f')
    sch = widgets.FloatSlider(description='chunk:', min=0,max=0.075,step=0.0005, 
                               value=wch, readout_format='.4f')

    butt = widgets.Button(description="Reset to defaults")
    startSim = widgets.Button(description="Start Simulation")

    def reset_to_defs(b):
        print('Weights reset to defaults')
        siit.value=wit
        sipf.value=wip
        spi.value=wpi
        sinh.value=wi
        sch.value=wch

    butt.on_click(reset_to_defs)

    ui2 = widgets.HBox([butt, startSim])
    ui = widgets.VBox([siit, sipf, spi, sinh, sch, ui2])

    def f(a, b, c, d, e):
        # print((a, b))
        # print((a, b))
        return(a)

    def simulate(b):
        print('Starting Simulation')
    
    def runSim(butt):
        w_intra_it = siit.value
        w_intra_pf = sipf.value
        w_pf_it = spi.value
        w_inh = sinh.value
        w_chunking = sch.value
        net.restore()
        net.run(500*ms, report='stdout')
        plotoutput('b: {}, c: {}, wmd: {}, wit: {}, wip: {}, wpi: {}, wi: {}, wc: {}'.format(b, c, wmod, w_intra_it, w_intra_pf, w_pf_it, w_inh, w_chunking))
        plot3(StM, 50)

    startSim.on_click(runSim)

    out = widgets.interactive_output(f, {'a': siit, 'b': sipf, 'c': spi,
                                        'd': sinh, 'e': sch})
    return(ui, out)
    #display(ui, out)
```
